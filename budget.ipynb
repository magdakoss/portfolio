# Hello, welcome to the monthly budget calculator!

## About
The Weekly Budget Calculator is a simple tool designed to help you track your monthly expenses and savings. By entering your income and expenses in various categories, you can get a clear picture of your monthly financial situation.

## How to Use
1. Run the program in your Python environment.
2. Enter your monthly income after taxes when prompted.
3. Enter your expenses for each category (e.g., food, entertainment, shopping, etc.).
4. The program will calculate and display your expenses and provide sugestions to your budget.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_positive_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value < 0:
                raise ValueError("Value must be non-negative.")
            return value
        except ValueError as e:
            print(f"Invalid input: {e}")


income = get_positive_float("How much do you make a month post taxes? $")
food = get_positive_float("How much do you spend on food a month, inlcuding grocories and take out? $")
entertainment = get_positive_float("How much do you spend on entertainment monthly? $")
shopping = get_positive_float("How much do you spend on pleasure shopping monthly? $")
transportation = get_positive_float("How much do you spend on transportation monthly? $")
utilities = get_positive_float("How much do you spend on utilities monthly? $")
house = get_positive_float("How much do you spend on housing monthly (mortgage/rent)? $")
other = get_positive_float("How much do you spend on other expenses monthly? If N/A type in 0 $")
```

In the 50/20/30 budget plan, 50% of your income should go to needs, 20% to savings, and 30% to wants.

```python
ideal_buget = np.array([50, 20, 30])
plt.pie(ideal_buget, 
        labels= ["Needs", "Savings", "Wants"], 
        startangle = 90, 
        colors = ["blue", "green", "red"],
        autopct='%1i%%',
        textprops={'color':"w"})
plt.title("Ideal Budget", color = "white")
plt.gcf().set_facecolor('none')
```

```python
total_expenses = food + entertainment + transportation + utilities + house + shopping + other
left_over = income - total_expenses
saving = left_over * 0.20

if total_expenses > income:
    print("Warning: Your expenses exceed your income! Consider reducing your spending.")
    
print(f"Your total monthly expenses are: ${total_expenses:.2f}")
print(f"Your left over income is: ${left_over:.2f}")
print(f"You should be at least saving ${saving:.2f} per month.")
```

```python
needs_percentage = ((food + transportation + utilities + house + other) / income) * 100
wants_percentage = ((entertainment + shopping) / income) * 100
left_over_percentage = (left_over / income) * 100
your_budget = np.array([needs_percentage, 20, left_over_percentage, wants_percentage])
plt.pie(your_budget,
        labels= ["Needs", "Savings", "Left Over Income", "Wants"],
        startangle = 90,
        colors = ["blue", "green", "black", "red"],
        autopct='%1.1f%%',
        textprops={'color':"w"})
plt.title("Your Budget", color = "white")
plt.gcf().set_facecolor('none')
```

```python
if wants_percentage > 30:
    print("You are spending more than 30% of your income on wants. Consider reducing your entertainment expenses.")

if needs_percentage > 50:
    print("You are spending more than 50% of your income on needs. Consider reducing your housing and transportation expenses.")
```
